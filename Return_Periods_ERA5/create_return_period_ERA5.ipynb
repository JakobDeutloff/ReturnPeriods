#### Imports

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

import os

import iris

import iris.quickplot as qplt
import iris.coord_categorisation as coord_cat
```

#### Check variables of the NetCDF file

```python
return_period = iris.load('precipitation-at-fixed-return-period_europe_ecad_30-year_5-yrs_1989-2018_v1.nc')
print (return_period[0])
```

#### Load data into an Iris cube (5-year return period as an example)

```python
return_period_path = 'precipitation-at-fixed-return-period_europe_ecad_30-year_5-yrs_1989-2018_v1.nc'
return_period = iris.load_cube(return_period_path, 'precipitation_5-return-period')
return_period.coord('latitude').guess_bounds()
return_period.coord('longitude').guess_bounds()
```

#### Plot some data

```python
qplt.pcolormesh(return_period)
plt.gca().coastlines()
```

#### Extract UK region

```python
# Create a constraint for the latitude and Longitude extents

UK_lat = iris.Constraint(latitude=lambda v: v > 50 and v <= 60 )
UK_lon = iris.Constraint(longitude=lambda v: v > -9 and v <= 3 )

return_period_UK =  return_period.extract(UK_lat & UK_lon) 
```

#### Plot the data

```python
print(return_period_UK)
qplt.pcolormesh(return_period_UK, vmin = 40, vmax = 55) # exclude vmin and vmax first, to check the range, then adjust it for visualisation purpose
plt.gca().coastlines()
```
